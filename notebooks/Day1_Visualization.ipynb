# Plot the Normal Distribution
Now you should be able to reproduce the following plot (taken from Wikipedia):
<img src='https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/720px-Normal_Distribution_PDF.svg.png'>

```python
from scipy.stats import norm
import math
```

```python
x_axis = np.arange(-5, 5, 0.001)
plt.plot(x_axis, norm.pdf(x_axis,0,np.sqrt(0.2)), 'b', label='µ=0 $\sigma^2$=0.2');
plt.plot(x_axis, norm.pdf(x_axis,0,np.sqrt(1)), 'r', label='µ=0 $\sigma^2$=1');
plt.plot(x_axis, norm.pdf(x_axis,0,np.sqrt(5)), 'y', label='µ=0 $\sigma^2$=5');
plt.plot(x_axis, norm.pdf(x_axis,-2,np.sqrt(0.5)), 'g', label='µ=-2 $\sigma^2$=0.5');
plt.xlabel('X');
plt.ylabel('$\phi_{\mu, \sigma^2}(X)$');
plt.axis([-5, 5, 0.0, 1.0])
#plt.xticks(np.arange(-5, 5, step=0.25))
plt.legend();
```
